```python
from datasets import load_dataset
from transformers import pipeline

nome_dataset = "google/fleurs"
idioma_dataset = "en_us"

dados = load_dataset(
    nome_dataset,
    name=idioma_dataset,
    split='train',
    streaming=True,
    trust_remote_code=True  # <- Adicionado aqui
)


modelo = "sanchit-gandhi/whisper-medium-fleurs-lang-id"
classificador = pipeline('audio-classification', model=modelo)


for linha in dados.take(5):
    predicao = classificador(linha['audio']['array'])
    print(predicao)
    print('-----')
```

```python
import sounddevice as sd

duracao = 10
taxa_amostragem = 16000
tamanho_vetor = int(duracao * taxa_amostragem)

gravacao = sd.rec(tamanho_vetor, samplerate=taxa_amostragem, channels=1)
sd.wait()

gravacao
```

```python
gravacao.shape
```

```python
gravacao = gravacao.ravel()
```

```python
gravacao.shape
```

```python
import IPython

IPython.display.Audio(data=gravacao.T, rate=taxa_amostragem)
```

```python
from transformers import pipeline

modelo = "sanchit-gandhi/whisper-medium-fleurs-lang-id"
classificador = pipeline('audio-classification', model=modelo)
```

```python
entrada_audio = gravacao.squeeze() 
resultado = classificador(entrada_audio)
print(resultado)
```
